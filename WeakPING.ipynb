# Ripples as inhibitory transients - model modification

Here, I attempt to modify Paola Malerba's (2016) model in order to drive it toward production of both a strong PING, and a weak PING (55 Hz range). Paola suggests modifying the parameters below independently, to start.

In order to do this: 
* set noise to 0 (in defaultparamsCA1.m)

* input DC (which affects the baseline level of excitability)

* increase the density of excitatory-excitatory connections
